```python
x=neural_network.iloc[:,:11]
x
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x['month']=le.fit_transform(x['month'])
x['day']=le.fit_transform(x['day'])
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler()
scaled_x=std_scaler.fit_transform(x)
scaled_x
scaled_x=pd.DataFrame(scaled_x)
scaled_x
```

```python
y=neural_network['size_category']
y
```

```python
y=neural_network['size_category']=le.fit_transform(neural_network['size_category'])
y
y=pd.DataFrame(y)
y
```

```python
y.dtypes
```

```python
scaled_x['y']=y
neural_network=scaled_x
neural_network
```

```python
from tensorflow import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation,Layer,Lambda
```

```python
neural_network.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python

train,test = train_test_split(neural_network,test_size = 0.3,random_state=42)
x_train = neural_network.iloc[0:361,:10]
y_train = neural_network.iloc[0:361,11:]
x_test= neural_network.iloc[361:,:10]
y_test =neural_network.iloc[361:,11:]
```

```python
x_train.shape  ,   y_train.shape  ,   x_test.shape
```

```python
x_train.std()
```

```python
x_train.mean()
```

```python
import tensorflow as tf
model = Sequential()
model.add(Dense(units=20,activation='relu'))
model.add(Dense(units=40,activation='tanh'))
model.add(Dense(units=10,activation='softmax'))                            
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x=x_train,y=y_train,epochs=500)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
result = model.evaluate(x=x_test,y=y_test)
result
```

```python
print('Model Accuracy : ',round(result[1],2))
print('Loss           : ',round(result[0],2))
```

```python
keras.metrics.binary_accuracy(y_test, y_pred, threshold=0.5)
```
